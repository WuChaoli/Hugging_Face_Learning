### 数据集的基本操作

#### 数据排序

```python
train_dataset[0:10]
```

```python
sorted_dataset = train_dataset.sort('label')
print(sorted_dataset['label'][0:10])
print(sorted_dataset['label'][-10:])
```

```python
shuffled_dataset1 = sorted_dataset.shuffle(seed=10)
shuffled_dataset2 = sorted_dataset.shuffle(seed=10)
print(shuffled_dataset1['label'][0:10])
print(shuffled_dataset2['label'][0:10])
print(shuffled_dataset1['label'][-10:])
```

#### 数据抽样

```python
selected_dataset = train_dataset.select([0,10,20,30,40])
print(selected_dataset)
```

#### 数据过滤

```python
def filter(data):
    return data['text'].startswith('非常不错') and data['label'] == 1
train_dataset.filter(filter)
```

```python
print(train_dataset[0])
```

#### 测试集划分

```python
temp_dataset = train_dataset.train_test_split(test_size=0.1, shuffle=True)
train_data, test_data = temp_dataset['train'], temp_dataset['test']
print(f'train_dataset: {len(train_data)}')
print(f'test_dataset: {len(test_data)}')
```

#### 数据分桶

```python
train_data_shards = train_dataset.shard(num_shards=4, index=0)
print(len(train_data_shards))
```

#### 字段操作

```python
# 重命名字段
train_dataset.rename_column('text', 'text_rename')
```

```python
# 删除字段
train_dataset.remove_columns(['text'])
```

```python
# 映射函数，主要用于处理数据
def mapping(data):
    data['text'] = 'My sentence ' + data['text']
    return data

mapping_data = train_dataset.map(mapping)

for snentence in mapping_data['text'][:5]:
    print(snentence)
```

```python
# 批处理加速
def mapping(datas):
    texts = datas['text']
    texts = ['My multiple sentence '+text for text in texts]
    datas['text'] = texts
    return datas

mapping_data = train_dataset.map(
    function=mapping,
    batched=True,
    batch_size=1000,
    num_proc=4
)

for snentence in mapping_data['text'][:5]:
    print(snentence)
```

```python
# 设置数据格式
train_dataset.set_format(
    type='torch',
    columns=['label'],
    output_all_columns=True
)
```
